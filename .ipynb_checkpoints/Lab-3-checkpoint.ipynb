입력 영상을 넣을 수 있는 placeholder를 선언한다. 이때 영상의 갯수는 추후 자유로이 넣을 수 있도록 None으로 선언한다.

```python
# 이미지 데이터 플레이스홀더
x = tf.placeholder("float", [None, 784])
```

Weight변수와 Bias변수에 대한 선언

```python
# 웨이트와 바이어스 변수
W = tf.Variable(tf.zeros([784,10]))
b = tf.Variable(tf.zeros([10]))
```

x와 W간의 행렬곱이 이후 softmax함수를 통과하는 연산을 선언함.

```python
# 모델 구현
y = tf.nn.softmax(tf.matmul(x,W) + b)

# 정답 레이블용 플레이스 홀더
y_ = tf.placeholder("float", [None,10])
```

Loss함수를 선언한다. 이번에는 cross entropy로 함

```python
# Loss 함수
cross_entropy = -tf.reduce_sum(y_*tf.log(y))
```

Optimizer를 Gradient Descent Optimizer로 놓고, 이 Optimizer로 cross_entropy를 줄이려는 operation을 정의 한다. 

```python
# 학습 오퍼레이션
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
```

```python
# 모든 변수 초기화
init = tf.initialize_all_variables()

sess = tf.Session()
sess.run(init)
```

ds.train.next_batch(10)는 학습용 영상과 거기에 해당되는 라벨을 10를 batch로 읽어들여서 학습용 영상 10는 batch_xs에 저장하고, 학습용 라벨 10개는 batch_ys에 저장한다.
batch_xs와 batch_ys를 입력시에 feeding하여 학습을 시작함.

```python
for i in range(1000):
  batch_xs, batch_ys = ds.train.next_batch(10)
  sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
```

학습이 끝난 후 y벡터에서 가장 큰 값이 나오는 원소의 위치와 y_벡터에서 가장 큰 값이 나오는 원소의 위치가 일치하는 y벡터,y_벡터 쌍이 몇개인지를 계산해서 correct_prediction에 넣는다.

```python
# 정답의 갯수를 구한다. 
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
```

이번에는 정답율을 구한다.

```python
# 정답율을 구한다. 
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))

print (sess.run(accuracy, feed_dict={x: ds.test.images, y_: ds.test.labels}))
```
